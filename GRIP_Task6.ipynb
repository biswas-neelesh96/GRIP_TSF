Everyone knows about the rising of Covid-19 cases in the world & and our country; and as of now, our country, India is in the second highest no of Covid-19 cases registered. Everyone knows how Covid-19 has been impacted in our stated but hardly anyone knows the same about their home districts. With the help of **Spreadsheets (Google Sheets)** and **Data Visualization (Python)** tools, I tried to make a detailed analysis about the Covid-19 cases in two districts of West Bengal:

1) *Uttar Dinajpur*: Where I was born & had grew up.

2) *Darjeeling:* Where I'm currently living now

I took the record of daily cases of both the districts from reliable sources like [www.covid19india.org](https://www.covid19india.org) & [wbhealth.gov.in](https://wbhealth.gov.in) and made two separate spreadsheets in the form of **.csv** files, one for each district. The results are concluded on basis of:

1) **No. of cases registered, recoveries & deaths** recorded in both the districts separately.

2) **Total no. of cases, Total no. of recoveries, Total no. of deaths & Active Cases** by comparing both the districts in a single graph respectively.

3) **Recovery Rate & Fatality Rate** by comparing both the districts in a single graph respectively.

4) Comparison of **No. of cases registered & No. of Active Cases in a single day.** (2.c & 5.c respectively)

5) No. of Covid-19 *New registered Cases & Active Cases* **for the last 30 days.**

6) **Overall Statistics of Covid-19 in both the districts** by separate graph for each district.

7) Expanding dataframes & Comparing the **Percentage of Covid-19 active cases (%) for both the districts** in a single graph. 

8) *Recovery Rate, Fatality Rate & Percentage of Active Cases* **for the last 7 days** in separate graphs. 

```
#import python libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
print('Setup Complete')
```

## 1.a) Creating a dataframe showing the cases of the last five days in Darjeeling followed by the Statistical Info

```
data_darj=pd.read_csv('Covid19(Darjeeling).csv', index_col='Date', parse_dates=True)
data_darj.tail()
```

```
data_darj.describe()
```

## 1.b) Creating a dataframe showing the cases of the last five days in Uttar Dinajpur followed by the Statistical Info

```
data_ud=pd.read_csv('Covid19(Uttar Dinajpur).csv', index_col='Date', parse_dates=True)
data_ud.tail()
```

```
data_ud.describe()
```

## 2.a) Covid-19 Cases registered in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Darjeeling as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_darj.index, data_darj['No of registered cases'])
```

## 2.b) Covid-19 Cases registered in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Uttar Dinajpur as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_ud.index, data_ud['No of registered cases'])
```

## 2.c) Covid-19 New Cases registered in a Single day at both the districts (Comparison)

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Covid-19 Cases Registered in a Single day in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Registered Cases in a Single day')
sns.lineplot(data=data_darj['No of registered cases'], label='Darjeeling', color='lime')
sns.lineplot(data=data_ud['No of registered cases'], label='Uttar Dinajpur', color='teal')
```

## 3.a) Covid-19 Cases recovered in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases recovered in a Single day at Darjeeling as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Cases recovered in a Single day')
plt.stem(data_darj.index, data_darj['No. of recoveries'],'g')
```

## 3.b) Covid-19 Cases recovered in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases recovered in a Single day at Uttar Dinajpur as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Cases recovered in a Single day')
plt.stem(data_ud.index, data_ud['No. of recoveries'],'g')
```

## 4.a) Covid-19 Deaths in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Deaths in a Single day at Darjeeling as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Deaths in a Single day')
plt.stem(data_darj.index, data_darj['No. of Deaths'],'#000000')
```

## 4.b) Covid-19 Deaths in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Deaths in a Single day at Uttar Dinajpur as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Deaths in a Single day')
plt.stem(data_ud.index, data_ud['No. of Deaths'],'#000000')
```

## 5.a) Covid-19 Active Cases in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Active Cases in a Single day at Darjeeling as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 New Cases in a Single day')
plt.stem(data_darj.index, data_darj['New Cases in a single day'],'m')
```

## 5.b) Covid-19 Active Cases in a Single day at Uttar Dinajpur


```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Active Cases in a Single day at Uttar Dinajpur as of 14th October 2020")
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 New Cases in a Single day')
plt.stem(data_ud.index, data_ud['New Cases in a single day'],'m')
```

## 5.c) Covid-19 Active Cases in a Single day at both the districts (Comparison)

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Covid-19 New Cases in a Single day in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Covid-19 New Cases in a Single day')
sns.lineplot(data=data_darj['New Cases in a single day'], label='Darjeeling', color='violet')
sns.lineplot(data=data_ud['New Cases in a single day'], label='Uttar Dinajpur', color='purple')
```

## 6) Total no of Covid-19 recoveries in both the districts (comparison)

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 recoveries in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Total Covid-19 Cases Recovered')
sns.lineplot(data=data_darj['Total Recoveries'], label='Darjeeling', color='navy')
sns.lineplot(data=data_ud['Total Recoveries'], label='Uttar Dinajpur', color='teal')
```

## 7) Total no of Covid-19 deaths in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Deaths in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Total no of Covid-19 Deaths')
sns.lineplot(data=data_darj['Total Deaths'], label='Darjeeling', color='black')
sns.lineplot(data=data_ud['Total Deaths'], label='Uttar Dinajpur', color='gray')
```

## 8) Total no of Covid-19 cases in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Cases in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Total Covid-19 Cases')
sns.lineplot(data=data_darj['Total Cases'], label='Darjeeling', color='green')
sns.lineplot(data=data_ud['Total Cases'], label='Uttar Dinajpur', color='red')
```

## 9) Active cases in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Active Cases in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('No of Active cases')
sns.lineplot(data=data_darj['No. of Active Cases'], label='Darjeeling', color='orange')
sns.lineplot(data=data_ud['No. of Active Cases'], label='Uttar Dinajpur', color='brown')
```

## 10) Covid-19 recovery rate in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Recovery Rate in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Recovery Rate (%)')
sns.lineplot(data=data_darj['Recovery Rate (%)'], label='Darjeeling', color='navy')
sns.lineplot(data=data_ud['Recovery Rate (%)'], label='Uttar Dinajpur', color='teal')
```

## 11) Covid-19 fatality rate in both the districts (comparison)

## 12) Creating separate dataframes & representing graphs which shows for the last 30 days:

a) No. of cases registered in a single day for both the districts.

b) No. of active cases in both the districts.




```
d1=data_darj.tail(30)
d2=data_ud.tail(30)
```

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Covid-19 Cases Registered in a Single day in Darjeeling & Uttar Dinajpur for the last 30 days')
plt.xlabel("Date")
plt.xticks(rotation=-45)
plt.ylabel('Covid-19 Registered Cases in a Single day')
sns.lineplot(data=d1['No of registered cases'], label='Darjeeling', color='lime')
sns.lineplot(data=d2['No of registered cases'], label='Uttar Dinajpur', color='teal')
```

```
plt.figure(figsize=(16,6))
plt.title('Active Cases in Darjeeling & Uttar Dinajpur for the last 30 days')
plt.xlabel("Date")
plt.xticks(rotation=-45)
plt.ylabel('No of Active cases')
sns.lineplot(data=d1['No. of Active Cases'], label='Darjeeling', color='orange')
sns.lineplot(data=d2['No. of Active Cases'], label='Uttar Dinajpur', color='brown')
```

## 13.a) Overall Statistics of Covid-19 in Darjeeling

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Overall Statistics of Covid-19 in Darjeeling as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel("No. of Covid-19 Cases")
sns.lineplot(data=data_darj['Total Cases'], label='Total Cases', color='red')
sns.lineplot(data=data_darj['Total Recoveries'], label='Total Recoveries', color='green')
sns.lineplot(data=data_darj['Total Deaths'], label='Total Deaths', color='black')
sns.lineplot(data=data_darj['No. of Active Cases'], label='Active Cases', color='orange')
```

## 13.b) Overall Statistics of Covid-19 in Uttar Dinajpur

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Overall Statistics of Covid-19 in Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel("No. of Covid-19 Cases")
sns.lineplot(data=data_ud['Total Cases'], label='Total Cases', color='red')
sns.lineplot(data=data_ud['Total Recoveries'], label='Total Recoveries', color='green')
sns.lineplot(data=data_ud['Total Deaths'], label='Total Deaths', color='black')
sns.lineplot(data=data_ud['No. of Active Cases'], label='Active Cases', color='orange')
```

## 14.a) Expanding the main dataframes by a column showing the percentage of active cases 

```
data_darj["Active Cases (%)"]=(data_darj["No. of Active Cases"]/data_darj["Total Cases"])*100
data_darj.tail()
```

```
data_ud["Active Cases (%)"]=(data_ud["No. of Active Cases"]/data_ud["Total Cases"])*100
data_ud.tail()
```

## 14.b) Percentage of Covid-19 Daily Active Cases (Comparison)

```
plt.figure(figsize=(16,6))
plt.title('Active Cases (%) in Darjeeling & Uttar Dinajpur as of 14th October 2020')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Active cases (%)')
sns.lineplot(data=data_darj['Active Cases (%)'], label='Darjeeling', color='orange')
sns.lineplot(data=data_ud['Active Cases (%)'], label='Uttar Dinajpur', color='brown')
```

## 15.a) Creating a dataframe showing the cases for the last 7 days

```
d3=data_darj.tail(7)
d4=data_ud.tail(7)
```

## 15.b) Plot showing the Recovery Rate for both the districts in the last 7 days

```
plt.figure(figsize=(16,6))
plt.title('Recovery Rate of cases in Darjeeling & Uttar Dinajpur for the last 7 days')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Percentage (%)')
sns.lineplot(data=d3['Recovery Rate (%)'], label='Darjeeling', color='navy')
sns.lineplot(data=d4['Recovery Rate (%)'], label='Uttar Dinajpur', color='teal')
```

## 15.c) Plot showing the Fatality Rate for both the districts in the last 7 days

```
plt.figure(figsize=(16,6))
plt.title('Fatality Rate of cases in Darjeeling & Uttar Dinajpur for the last 7 days')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Percentage (%)')
sns.lineplot(data=d3['Fatality Rate (%)'], label='Darjeeling', color='black')
sns.lineplot(data=d4['Fatality Rate (%)'], label='Uttar Dinajpur', color='gray')
```

## 15.d) Plot showing the % of Active Cases for both the districts in the last 7 days

```
plt.figure(figsize=(16,6))
plt.title('Active % of cases in Darjeeling & Uttar Dinajpur for the last 7 days')
plt.xlabel("Date")
#plt.xticks(rotation=-45)
plt.ylabel('Percentage (%)')
sns.lineplot(data=d3['Active Cases (%)'], label='Active % (Darjeeling)', color='orange')
sns.lineplot(data=d4['Active Cases (%)'], label='Active % (Uttar Dinajpur)', color='brown')
```
